```python
marks=input("Enter student marks")
marks=int(marks)
if marks>45:
  print("student pass in subject grade")
if marks>=45 and marks<=55:
  print("Pass")
elif marks<45:
  print("Fail")
if marks>=80:
  print("Grade C")
if marks>60 and marks<80:
  print("Grade D")
```
